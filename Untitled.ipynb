### since we have limited dataset, let's use vinzorisation to treat outliers

```python
# List of numerical columns to handle outliers
numerical_columns = X_train.select_dtypes(include=['float64', 'int64']).columns

# Loop through each numerical column
for column in numerical_columns:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = X_train[column].quantile(0.25)
    Q3 = X_train[column].quantile(0.75)
    
    # Compute IQR
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap the outliers
    X_train[column] = X_train[column].clip(lower=lower_bound, upper=upper_bound)

# Check if outliers are capped
print(X_train.describe())
```

## Handling the Imbalance Nature of the Dataset

#### The Synthetic Minority Oversampling Technique (SMOTE) is an oversampling method used to handle imbalanced datasets by generating synthetic samples for the minority class. Unlike traditional oversampling, which duplicates existing minority samples, SMOTE creates new samples by interpolating between a data point and its k-nearest neighbors in the feature space. This approach helps the model learn better decision boundaries and reduces overfitting, a common problem with duplicated samples

#### Basic advantage of using SMOTE other over sampling techniques is that it generates synthetic samples instead of simply duplicating existing minority class instances. This approach helps the model learn more diverse patterns within the minority class, which leads to better decision boundaries and reduces the risk of overfitting


```python

from imblearn.over_sampling import SMOTE

# Apply SMOTE to the training set
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Create subplots for side-by-side visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Count plot before SMOTE
sns.countplot(x=y_train, ax=axes[0])
axes[0].set_title("Before SMOTE")
axes[0].set_xlabel("Status")
axes[0].set_ylabel("Count")

# Count plot after SMOTE
sns.countplot(x=y_train_resampled, ax=axes[1])
axes[1].set_title("After SMOTE")
axes[1].set_xlabel("Status")

# Adjust layout and show the plots
plt.tight_layout()
plt.show()

```

## Machine Learning Model Development: Placement Prediction with Support Vector Machine

```python
# Train SVM model
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define the parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],  # Test different regularization strengths
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': [1, 0.1, 0.01, 0.001]  # Test different gamma values
}

# Create the GridSearchCV object
grid_search = GridSearchCV(SVC(probability=True, random_state=42), param_grid, cv=5, scoring='accuracy')

# Fit the model to the data
grid_search.fit(X_train_scaled, y_train)

# Output the best parameters and accuracy
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Training Accuracy:", grid_search.best_score_)

# Use the best model to make predictions
best_model = grid_search.best_estimator_
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Define the parameter distributions
param_dist = {
    'C': np.logspace(-3, 3, 10),  # Exponential scale for `C`
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'gamma': np.logspace(-4, 1, 10)  # Exponential scale for `gamma`
}

# Create the RandomizedSearchCV object
random_search = RandomizedSearchCV(SVC(probability=True, random_state=42), param_dist, n_iter=20, cv=5, scoring='accuracy', random_state=42)

# Fit the model to the data
random_search.fit(X_train_scaled, y_train)

# Output the best parameters and accuracy
print("Best Hyperparameters:", random_search.best_params_)
print("Best Training Accuracy:", random_search.best_score_)

# Use the best model to make predictions
best_model = random_search.best_estimator_
```

```python
from sklearn.metrics import accuracy_score

# Predict on the training set
train_preds = best_model.predict(X_train_scaled)

# Calculate accuracy for the training set
train_accuracy = accuracy_score(y_train, train_preds)
print(f"Training Set Accuracy: {train_accuracy:.2f}")
```

```python
# Predict on the test set
test_preds = best_model.predict(X_test_scaled)

# Calculate accuracy for the test set
test_accuracy = accuracy_score(y_test, test_preds)
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

#### The model achieved a training set accuracy of 90% and a test set accuracy of 90%. Since both accuracies are identical, it suggests that the model is generalized well to unseen data and is not likely overfitting. Overfitting typically occurs when a model performs significantly better on the training data than on the test data. In this case, with matching accuracy scores, it suggests that the model effectively captures the underlying patterns in the data without memorizing the training examples. Thus, there is no overfitting based on these results.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate confusion matrices for training and test sets
train_cm = confusion_matrix(y_train, train_preds)
test_cm = confusion_matrix(y_test, test_preds)

# Plot both confusion matrices side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Training set confusion matrix
ConfusionMatrixDisplay(train_cm).plot(ax=axes[0], colorbar=False)
axes[0].set_title("Confusion Matrix - Training Set")

# Test set confusion matrix
ConfusionMatrixDisplay(test_cm).plot(ax=axes[1], colorbar=False)
axes[1].set_title("Confusion Matrix - Test Set")

plt.tight_layout()
plt.show()
```

####
**Training Set Confusion Matrix**<br>

**Test Set Confusion Matrix**<br>
The test set confusion matrix shows similarly good results, with the model correctly predicting 26 instances of class 0 and 29 instances of class 1, alongside 3 false negatives and 3 false positives.
```python
from sklearn.metrics import classification_report

# Generate and print classification report for the test set
report = classification_report(y_test, test_preds)
print("Classification Report for Test Set:")
print(report)
```

The classification report provides key metrics for evaluating the model's performance. The precision indicates how many of the predicted positives are true positives, while recall reflects how many actual positives were correctly identified. The f1-score is the harmonic mean of precision and recall, balancing both metrics. In this report, the precision, recall, and f1-scores are all high (around 0.900.91) for both classes (0 and 1), indicating the model is effective in predicting both categories. The support column shows the number of actual instances for each class. The accuracy represents the overall performance of the model, which is 90% in this case.

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Calculate the predicted probabilities for the test set
y_test_proba = best_model.predict_proba(X_test_scaled)[:, 1]

# Generate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_test_proba)

# Calculate the AUC
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--', label='Random Guess')

plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()

print(f"Area Under the Curve (AUC): {roc_auc:.2f}")
```

### ROC Curve and AUC Value
The Receiver Operating Characteristic (ROC) curve is a graphical representation of a model's diagnostic ability, plotting the true positive rate (sensitivity) against the false positive rate. The Area Under the Curve (AUC) value of 0.93 indicates excellent model performance, as it suggests a high likelihood of correctly distinguishing between the positive and negative classes.

### Interpretation
An AUC of 0.93 signifies that the model has a strong ability to differentiate between classes, with a curve significantly above the random guess line. This suggests that the estimator is effective, as it achieves high sensitivity with relatively low false positive rates, making it a reliable choice for classification tasks.
