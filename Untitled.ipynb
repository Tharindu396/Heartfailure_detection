# Feature selection 

We can only coorelation matrix as all the data is in numerical form

```python
from scipy.stats import chi2_contingency
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Chi-Square Test (for categorical variables)
categorical_columns = ["anaemia", "diabetes", "high_blood_pressure", "sex", "smoking", "DEATH_EVENT"]
chi2_results = {}

for col in categorical_columns:
    if col != "DEATH_EVENT":  # Exclude target variable as the dependent variable
        contingency_table = pd.crosstab(df[col], df["DEATH_EVENT"])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        chi2_results[col] = {"Chi2 Statistic": chi2, "p-value": p}

print("Chi-Square Test Results:")
print(pd.DataFrame(chi2_results).T)  # Display results in a readable format

# Correlation Matrix (for numerical variables)
# Exclude categorical variables (binary variables are okay to include but interpret cautiously)
numerical_columns = ["age", "creatinine_phosphokinase", "ejection_fraction",
                     "platelets", "serum_creatinine", "serum_sodium", "time"]

correlation_matrix = df[numerical_columns + ["DEATH_EVENT"]].corr()

# Plot the heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Step 1: Chi-Square Test Feature Selection (p-value < 0.05 indicates significance)
# Features with significant chi-square p-values
significant_features_chi2 = [col for col, result in chi2_results.items() if result["p-value"] < 0.5]
print(f"Significant features based on Chi-Square Test: {significant_features_chi2}")

# Step 2: Correlation Matrix - Remove highly correlated features (correlation > 0.85)
correlation_threshold = 0.2
correlated_features = set()

# Identify correlated features (correlation > 0.85)
correlation_matrix = df[numerical_columns + ["DEATH_EVENT"]].corr()

# Iterate through the correlation matrix and identify pairs of correlated features
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > correlation_threshold:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)

print(f"Highly correlated features (correlation > 0.85): {correlated_features}")

# Step 3: Combine results from Chi-Square Test and Correlation Matrix
# Combine significant features from chi-square and drop highly correlated features
selected_features = list(set(significant_features_chi2) - correlated_features)
print(f"Final selected features after combining both tests: {selected_features}")
```
