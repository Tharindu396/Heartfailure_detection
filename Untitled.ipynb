**Q:** Can you recognize that the dataset is imbalnaced? Mention three problems of imbalnced dataset may cause during the machine learning model traning. <br>




#### From the bar chart in the image, it seems that the counts for the two categories (0 and 1) in the dataset are relatively close to each other. Generally, a dataset is considered imbalanced when the difference between the counts of the target classes is significant.
1.**Biased Predictions**<br>
 The model may become biased toward the majority class, as it will focus more on predicting the more frequent class and neglect the minority class. This can result in poor performance for predicting the minority class.<br>
2.**Poor Generalization**<br> 
The model may not generalize well to real-world data, where both classes are likely to appear more evenly distributed. This leads to a model that performs well on the majority class but poorly on the minority class.<br>
3.**Incorrect Performance Metrics** <br> 
Standard metrics like accuracy may not reflect the true performance of the model, as a high accuracy can be achieved by simply predicting the majority class most of the time. This can hide issues such as high false positives or false negatives for the minority class.<br>

#### Checking for Outliers

```python
numerical_columns = X_train.select_dtypes(include=['float64', 'int64']).columns

# Plot boxplots for each numerical column in the training set
for column in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=X_train[column])
    plt.title(f"Boxplot of {column} (Training Set)")
    plt.xlabel(column)
    plt.show()
```

### since we have limited dataset, let's use vinzorisation to treat outliers

```python
# List of numerical columns to handle outliers
numerical_columns = X_train.select_dtypes(include=['float64', 'int64']).columns

# Loop through each numerical column
for column in numerical_columns:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = X_train[column].quantile(0.25)
    Q3 = X_train[column].quantile(0.75)
    
    # Compute IQR
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap the outliers
    X_train[column] = X_train[column].clip(lower=lower_bound, upper=upper_bound)

# Check if outliers are capped
print(X_train.describe())
```

## Handling the Imbalance Nature of the Dataset

#### The Synthetic Minority Oversampling Technique (SMOTE) is an oversampling method used to handle imbalanced datasets by generating synthetic samples for the minority class. Unlike traditional oversampling, which duplicates existing minority samples, SMOTE creates new samples by interpolating between a data point and its k-nearest neighbors in the feature space. This approach helps the model learn better decision boundaries and reduces overfitting, a common problem with duplicated samples

#### Basic advantage of using SMOTE other over sampling techniques is that it generates synthetic samples instead of simply duplicating existing minority class instances. This approach helps the model learn more diverse patterns within the minority class, which leads to better decision boundaries and reduces the risk of overfitting


```python

from imblearn.over_sampling import SMOTE

# Apply SMOTE to the training set
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Create subplots for side-by-side visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Count plot before SMOTE
sns.countplot(x=y_train, ax=axes[0])
axes[0].set_title("Before SMOTE")
axes[0].set_xlabel("Status")
axes[0].set_ylabel("Count")

# Count plot after SMOTE
sns.countplot(x=y_train_resampled, ax=axes[1])
axes[1].set_title("After SMOTE")
axes[1].set_xlabel("Status")

# Adjust layout and show the plots
plt.tight_layout()
plt.show()

```

As it can be seen from the above plot the the SMOTE has balanced the traning dataset by oversampling the minority class. <br>

## Machine Learning Model Development: Placement Prediction with Support Vector Machine

```python
# Train SVM model
svm_model = SVC(kernel="rbf", probability=True, random_state=42)
svm_model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Predict on the training set
train_preds = svm_model.predict(X_train_scaled)

# Calculate accuracy for the training set
train_accuracy = accuracy_score(y_train, train_preds)
print(f"Training Set Accuracy: {train_accuracy:.2f}")
```

```python
# Predict on the test set
test_preds = svm_model.predict(X_test_scaled)

# Calculate accuracy for the test set
test_accuracy = accuracy_score(y_test, test_preds)
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

**Q:** Comment on the accuracies obtained above. Do you think this model is overfitting or not?

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate confusion matrices for training and test sets
train_cm = confusion_matrix(y_train, train_preds)
test_cm = confusion_matrix(y_test, test_preds)

# Plot both confusion matrices side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Training set confusion matrix
ConfusionMatrixDisplay(train_cm).plot(ax=axes[0], colorbar=False)
axes[0].set_title("Confusion Matrix - Training Set")

# Test set confusion matrix
ConfusionMatrixDisplay(test_cm).plot(ax=axes[1], colorbar=False)
axes[1].set_title("Confusion Matrix - Test Set")

plt.tight_layout()
plt.show()
```

**Q:** Comment about the obtained confusion matrices.

```python
from sklearn.metrics import classification_report

# Generate and print classification report for the test set
report = classification_report(y_test, test_preds)
print("Classification Report for Test Set:")
print(report)
```

**Q:** Comment on the results obtained with classfication report. Explain the different parameters you can observe in the report.

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Calculate the predicted probabilities for the test set
y_test_proba = svm_model.predict_proba(X_test_scaled)[:, 1]

# Generate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_test_proba)

# Calculate the AUC
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--', label='Random Guess')

plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()

print(f"Area Under the Curve (AUC): {roc_auc:.2f}")
```
